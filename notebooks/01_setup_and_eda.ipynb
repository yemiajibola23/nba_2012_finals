```python
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.static import teams
import time

nba_teams = teams.get_teams()
heat = [team for team in nba_teams if team['abbreviation'] == 'MIA'][0]
thunder = [team for team in nba_teams if team['abbreviation'] == 'OKC'][0]

print("Heat team info:", heat)
print("Thunder team info:", thunder)

```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=heat['id'], season_nullable='2011-12', season_type_nullable='Playoffs')

heat_playoffs = gamefinder.get_data_frames()[0]

print("Shape", heat_playoffs.shape)
print("\nColumns:", heat_playoffs.columns)
print("\nAll Heat playoff games:")
heat_playoffs
```
